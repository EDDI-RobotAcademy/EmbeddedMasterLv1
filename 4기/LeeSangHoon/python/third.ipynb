```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
## __init__ 은 생성자이다.

    def print(self):
        print("이름은 {0}, 나이는 {1}".format(self.name, self.age))
        
p = Person("고고싱", 77)

## 클래스 이름과 같은 함수가 호출되면 생성자 호출이다.


# print(p.__dict__)
p.print()
p.__dict__
```

```python
class Student(Person): ## 파이썬은 상속을 class 안에 변수로 적어 상속받는 개념 
    def __init__(self, name, age, major):
        super(Student, self).__init__(name, age)
        ## 상속받은 클래스의 name과 age를 받아오기 위함 
        self.major = major
        ## 이거는 Student class 고유의 생성자
    def printStudent(self):
        print("Student - name: {0}, age: {1}, major: {2}".format(self.name, self.age, self.major))
        
s = Student("고고싱", 77, "물리학")

print(s.__dict__)
```

```python
import time
import multiprocessing

def count_down(x):
    while True:
        if x == 0:
            break
            
        print("카운트 다운 ... %d" % x)
        x -= 1
        time.sleep(1)
        
p = multiprocessing.Process(target=count_down, args=(5, ))
p.start()
```

```python
):import threading
import os

def info():
    print('모듈명: ', __name__)
    print('프로세스 id: ', os.getpid())
    print('스레드 id: ', threading.get_ident())
    print('이런식으로 많이 집어넣으면 Context Switching도 확인할 수 있음')
    
def print_hello():
    print('안녕')
    info()
    time.sleep(1)
    print('잠자다 일어난다\n')
    
if __name__ == '__main__':
    # 현재 시간 기록
    start = time.perf_counter()
    threads = []
    
    # 스레드 5개 생성
    # 실제 컴퓨터 성능에 따라서 이 부분에 많이 혼성될 수도 있고 덜 혼성될 수도 있음
    # 컴퓨터가 압도적으로 좋거나 압도적으로 후지면 잘 혼성되지 않는다.
    for _ in range(100):
        t = threading.Thread(target = print_hello)
        t.start()
        threads.append(t)
        
    # 스레드 구동
    for thread in threads:
        thread.join()
        
    # 현재 시간 기록(현재 시간 - 이전 시간) -> 걸린 시간
    finish = time.perf_counter()
    
    print('종료: ', str(round(finish - start, 2)) + '초')
```

```python
import threading

x = 0

def increment_global():
    global x
    x += 1
    
def thread_processing():
    for _ in range(1000000):
        increment_global()
        
def thread_main():
    global x
    x = 0
    
    t1 = threading.Thread(target = thread_processing)
    t2 = threading.Thread(target = thread_processing)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
for i in range(10):
    thread_main()
    print("{0} 번째 반복 이후 x = {1}".format(i, x))
```

```python
import threading

lock = threading.Lock()
x = 0

def increment_global():
    global x
    x += 1
    
def thread_processing():
    for _ in range(1000000):
        lock.acquire()
        increment_global()
        lock.release()
        
def thread_main():
    global x
    x = 0
    
    t1 = threading.Thread(target = thread_processing)
    t2 = threading.Thread(target = thread_processing)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
start = time.perf_counter()
    
for i in range(10):
    thread_main()
    print("{0} 번째 반복 이후 x = {1}".format(i, x))
    
finish = time.perf_counter()
print('종료: ', str(round(finish - start, 2)) + '초')
```
